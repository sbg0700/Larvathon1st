```python

# ---------------------------------------------------------------
#  제    목 :국민건강영양조사(KNHANES) 데이터를 기반으로
#            당뇨병 진단을위한 머신러닝 프로젝트
#  기초자로 : 2010년부터 2023년까지의 (14년치)
#  날    짜 : 20251024 금요일 2PM
#  작    성 : KWON
#  팀    명 : lavarton
#  환    경 : 파이선, 코랩
#-----------------------------------------------------------------


# 1. '재료' (라이브러리) 가져오기
# 우리가 요리를 하려면 칼, 도마, 냄비 같은 도구가 필요하듯이,
# 파이썬에서 데이터를 분석하고 모델을 만들려면 특별한 '도구 상자'를 가져와야 해요.

import pandas as pd     # 엑셀처럼 표 모양의 데이터(DataFrame)를 다루는 가장 중요한 도구
import numpy as np      # 복잡한 숫자 계산을 빠르고 쉽게 해주는 도구
import matplotlib.pyplot as plt # 그래프를 그려서 결과를 눈으로 보여주는 도구
import seaborn as sns   # Matplotlib보다 더 예쁘고 멋진 그래프를 쉽게 그릴 수 있는 도구
import platform         # 지금 코드가 돌아가고 있는 컴퓨터의 정보를 알려주는 도구 (필수는 아님)

# '랜덤 포레스트'라는 특별한 '예측 기계'를 만드는 데 필요한 부품들
from sklearn.model_selection import train_test_split # 데이터를 '연습용'과 '시험용'으로 나누는 도구
from sklearn.preprocessing import LabelEncoder, StandardScaler # 데이터를 모델이 이해하기 좋게 다듬는 도구
from sklearn.ensemble import RandomForestClassifier # '랜덤 포레스트' 예측 기계 자체
from sklearn.metrics import classification_report, roc_auc_score # 예측 기계가 얼마나 잘 맞췄는지 점수를 매기는 도구


# 2. 그래프 글꼴 설정 (폰트 오류 방지)
# 그래프의 글자들이 깨지거나 네모로 나오지 않도록 설정해요.
# 특히 마이너스(-) 부호가 깨지는 것을 막아주는 중요한 설정이에요.
plt.rcParams['font.family'] = 'DejaVu Sans' # 영문 폰트가 잘 보이도록 가장 기본적인 폰트를 지정
plt.rcParams['axes.unicode_minus'] = False  # 마이너스(-) 기호가 네모로 깨지는 것을 방지


# 3. 연도별 건강 데이터 합치기
# 2010년부터 2023년까지 매년 따로 저장된 건강 검진 기록부 파일을 하나로 모으는 작업이에요.
years = range(2010, 2024) # 2010년부터 2023년까지의 숫자 목록 (14년치)
df_list = [] # 합칠 데이터들을 잠시 담아둘 빈 상자

for year in years:
    file = f"HN{year}_ALL.csv" # 파일 이름 만들기 (예: HN2010_ALL.csv)
    try:
        temp = pd.read_csv(file) # 파일을 읽어와서 표(데이터프레임)로 만듦
        temp['year'] = year      # 이 데이터가 몇 년도 것인지 표시하는 'year'라는 새 칸을 추가
        df_list.append(temp)     # 만든 표를 빈 상자에 넣음
    except Exception as e:
        # 혹시 파일이 없거나 열리지 않으면 '에러 메시지'를 출력
        print(f"{file} error: {e}")

# 상자 안에 모은 모든 표들을 위에서 아래로 주욱 붙여서 하나의 큰 표(df)로 완성!
df = pd.concat(df_list, ignore_index=True)


# 4. '당뇨병 예측'에 사용할 중요한 정보 (특성) 선택
# 모든 건강 기록 중에서 당뇨병 예측에 꼭 필요할 것 같은 항목들만 고르는 거예요.
features = [
    'HE_HbA1c',      # 당화혈색소: 3개월 동안의 평균 혈당을 알려주는 지표
    'HE_glu',        # 공복혈당: 밥 안 먹고 잰 혈당 수치
    'HE_BMI',        # 체질량지수: 키와 몸무게로 계산한 비만도
    'HE_wc',         # 허리둘레: 배가 얼마나 나왔는지
    'HE_sbp',        # 수축기 혈압: 심장이 뛸 때의 높은 혈압
    'HE_dbp',        # 이완기 혈압: 심장이 쉴 때의 낮은 혈압
    'mod_d',         # 중등도 운동 여부: 숨이 찰 정도의 운동을 하는지
    'sm_presnt',     # 흡연 여부: 담배를 피우는지
    'HE_DMfh1',      # 당뇨병 가족력 (부모): 부모님이 당뇨병이 있는지
    'HE_DMfh2',      # 당뇨병 가족력 (형제자매)
    'HE_DMfh3',      # 당뇨병 가족력 (기타 친척)
    'age',           # 나이
    'sex',           # 성별 (남자/여자)
    'region',        # 지역 (어디에 사는지)
    'year'           # 몇 년도 검진 데이터인지
]
target = 'DI1_dg'  # 타겟 변수: 우리가 예측하고 싶은 '정답' (당뇨병 진단: 0=아님, 1=맞음)

# 선택한 특성들과 정답('DI1_dg')만 따로 복사해서 새로운 표를 만듦
df_model = df[features + [target]].copy()


# 5. 예측 기계가 이해할 수 있도록 데이터 다듬기 (전처리)
# 데이터에 있는 지저분하거나 불완전한 부분을 청소하고 정리하는 과정이에요.

# 공백(' ')을 '모르는 값(NaN)'으로 바꿔줘요.
df_model.replace(' ', np.nan, inplace=True)

# '성별', '지역' 같은 글자 정보(범주형)를 숫자로 바꿔주는 작업 (레이블 인코딩)
# 예를 들어, '남자'는 0, '여자'는 1 처럼 숫자로 바꿔 모델이 이해하게 해요.
for col in ['sex', 'region']:
    df_model[col] = LabelEncoder().fit_transform(df_model[col].astype(str))

# 모든 데이터를 강제로 '숫자'로 바꾸고, 숫자가 아닌 것은 '모르는 값(NaN)'으로 만듦
for col in features + [target]:
    df_model[col] = pd.to_numeric(df_model[col], errors='coerce')

# **결측치(모르는 값) 처리:**
missing_ratio = df_model.isnull().mean() # 각 항목별로 '모르는 값'이 전체의 몇 %인지 계산
# '모르는 값'이 전체의 30%보다 적은 항목들만 '괜찮은 특성'으로 인정하고 목록을 만듦
valid_features = missing_ratio[missing_ratio < 0.3].index.tolist()
if target not in valid_features:
    valid_features.append(target)

# '괜찮은 특성'만 남기고, '모르는 값'이 있는 줄(행)은 아예 지워버려요. (전체 데이터를 다 사용하지 못할 수 있음)
df_model = df_model[valid_features].dropna()
# 정답('DI1_dg')이 0 또는 1인 데이터만 사용 (당뇨병 예측 문제에 집중)
df_model = df_model[df_model[target].isin([0, 1])]

# 최종적으로 사용할 특성(X)과 정답(y)을 분리
features = [col for col in valid_features if col != target]
X = df_model[features] # 예측에 사용할 특성들 (문제지)
y = df_model[target].astype(int) # 예측하고 싶은 정답 (정답지)

# **데이터 스케일링 (표준화):**
# 모든 특성의 크기를 비슷하게 맞춰주는 작업이에요. (예: '나이(50)'와 '혈당(150)'의 숫자가 너무 차이나면 안 되니까)
# '스탠다드 스케일러(StandardScaler)'라는 도구를 사용해서 평균 0, 표준편차 1로 맞춰요.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# **데이터 분할:**
# 예측 기계를 훈련시킬 '연습용' 데이터(80%)와 성능을 테스트할 '시험용' 데이터(20%)로 나눔
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)


# 6. 예측 기계 훈련 및 평가 (Train and evaluate model)
# **모델 기법: 랜덤 포레스트 (Random Forest)**
# 🌳 **랜덤 포레스트란?** # - '포레스트(숲)'는 여러 개의 '트리(나무)'로 이루어져 있어요.
# - **트리(나무) 하나:** '당뇨병일까 아닐까?'를 예측하는 단순한 '예/아니오' 질문으로 이루어진 의사결정 나무예요.
# - **랜덤 포레스트:** 수백, 수천 개의 **랜덤하게 만들어진** 작은 '나무'들에게 같은 문제를 풀게 한 다음,
#   가장 많은 '나무'들이 내놓은 답을 최종 정답으로 결정하는 방식이에요. (투표!)
# - **장점:** 여러 나무의 의견을 모으기 때문에, 단 하나의 나무가 실수하더라도 최종 결과는 훨씬 **정확하고 안정적**이 됩니다.
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train) # 연습용 데이터(X_train, y_train)를 사용해서 모델(기계)을 훈련시킴

# 시험용 데이터로 예측 수행
y_pred = model.predict(X_test) # '시험용 문제지(X_test)'로 예측 기계가 답을 냄
y_proba = model.predict_proba(X_test)[:, 1] # '당뇨병(1)'일 확률이 몇 %인지 계산

# **평가:** 예측 기계가 얼마나 잘했는지 점수 매기기
print("📊 분류 보고서:")
print(classification_report(y_test, y_pred)) # 정확도(Accuracy), 정밀도(Precision), 재현율(Recall) 등 자세한 점수표 출력

try:
    # AUC 점수: 모델이 정답과 오답을 얼마나 잘 구별하는지 나타내는 점수 (1에 가까울수록 최고)
    auc_score = roc_auc_score(y_test, y_proba)
    print("AUC Score:", auc_score)
except ValueError as e:
    print("⚠️ AUC 계산 오류:", e)


# 7. 어떤 정보가 가장 중요했을까? (특성 중요도 시각화)
# '당뇨병 예측'을 할 때, 어떤 건강 항목(특성)이 가장 중요하게 작용했는지 보여주는 그래프예요.
feature_labels = {
    'HE_HbA1c': 'Hemoglobin A1c (당화혈색소)',
    'HE_glu': 'Fasting Glucose (공복혈당)',
    # ... (생략: 나머지 항목들도 위와 같이 주석 처리 가능)
    'year': 'Survey Year (조사 연도)'
}

importances = model.feature_importances_ # 랜덤 포레스트 모델이 계산한 각 특성의 중요도 점수
labels_eng = [feature_labels.get(f, f) for f in features] # 그래프에 표시할 영어/한국어 레이블

plt.figure(figsize=(10, 6))
# 중요도가 높은 순서대로 정렬해서 막대 그래프를 그립니다.
sorted_idx = importances.argsort()
sns.barplot(x=importances[sorted_idx], y=[labels_eng[i] for i in sorted_idx])
plt.title("Feature Importance for Diabetes Prediction (당뇨병 예측 중요도)")
plt.xlabel("Importance (중요도)")
plt.ylabel("Feature (특성)")
plt.tight_layout()
plt.show()


# 8. 연도별 당뇨병 유병률 추세 보기
# 당뇨병에 걸린 사람의 비율(유병률)이 시간이 지나면서 어떻게 변했는지 그래프로 확인해요.
if 'year' in df_model.columns:
    # 'year'별로 당뇨병 진단(target)의 평균을 계산 = 연도별 유병률
    trend = df_model.groupby('year')[target].mean()
    plt.figure(figsize=(10, 5))
    # 선 그래프를 그려서 추세를 한 눈에 보여줍니다.
    plt.plot(trend.index, trend.values, marker='o', linestyle='-', color='darkred')
    plt.title("Diabetes Prevalence Over Time (연도별 당뇨병 유병률)")
    plt.xlabel("Survey Year (조사 연도)")
    plt.ylabel("Prevalence Rate (유병률)")
    plt.grid(True) # 그래프에 격자 무늬를 넣어 보기 쉽게 함
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ 'year' 컬럼이 없어 추세 분석을 건너뜁니다.")
```
